```python
import pandas as pd
```

```python
app_data=pd.read_csv('googleplaystore.csv')
app_data
```

```python
app_data.dtypes
```

Cleaning the 1st Dataset

```python
app_data_clean=app_data.dropna(axis=0,how='any')
app_data_clean
```

```python
app_reviews=pd.read_csv('googleplaystore_user_reviews.csv')
app_reviews
```

```python
app_reviews.dtypes
```

Cleaning the 2nd Dataset.

```python
app_reviews_clean=app_reviews.dropna(axis=0,how='any')
app_reviews_clean
```

Merging both the cleaned datasets by using the common attribute 'App' in both the tables.

```python
joined_data=pd.merge(app_data_clean,app_reviews_clean,on='App',how='inner')
joined_data
```
