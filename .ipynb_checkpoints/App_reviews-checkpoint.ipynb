The goal of this project is to predict User Ratings of a google play store application based on predictor variables. 
We are using 3 datasets:
Google Play Store: This dataset is downloaded from Kaggle, which has information of all the apps on the google play store
Google Play Store User Reviews: This dataset has data about the user review of the apps on the google store
Ratings result: This dataset is created by us, which maps the rating scale and to its labels.

```python
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Importing all the datasets
app = pd.read_csv('googleplaystore.csv')
app_reviews = pd.read_csv('googleplaystore_user_reviews.csv')
app_ratings = pd.read_csv('Rating_result.csv')
app = app.rename(columns={"Content Rating": "content_rating","Current Ver": "current_ver", "Last Updated": "last_updated"})
```

So apart from Rating all others could be categorical.
Reviews, Size are continous variables and would not considered as categorical variables
We would check the categorical columns for number of distinct values -- a single unique value or an extremely high number would be bad.

```python
cat_columns = ['App', 'Category', 'Reviews', 'Size', 'Installs', 'Type',
       'Price', 'content_rating', 'Genres', 'last_updated', 'current_ver',
       'Android Ver']
# Use apply to check the data type and call the .nunique() method on each column
app[cat_columns].apply(lambda col: col.nunique())
```

We observe that App, Current Version, Last Updated have an extremely high number of distinct values.
So we are going to drop these columns in our final model.

```python
#Data cleaning
#Dropping columns which are not useful for determining the response variable
app = app.drop(['current_ver','last_updated','Android Ver','content_rating'] , axis = 1)
app_reviews = app_reviews.drop(['Translated_Review','Sentiment_Polarity','Sentiment_Subjectivity'] , axis = 1)
```

```python
app.isnull().sum(axis=0).reset_index()
```

Cleaning the 1st Dataset

```python
app=app.dropna(axis=0,how='any')
app
```

```python
app_reviews.isnull().sum(axis=0).reset_index()
```

```python
app_reviews.dtypes
```

Cleaning the 2nd Dataset.

```python
app_reviews=app_reviews.dropna(axis=0,how='any')
app_reviews
```

Merging both the cleaned datasets by using the common attribute 'App' in both the tables.

```python
#Joining the datasets
joined_data = pd.merge(app,app_reviews,on='App',how='inner')
joined_data
```

```python
joined_data['Price'].unique()
```

```python
# Function to remove $ from Price column
def remove(price):
    x = 0
    if price != '0':
        x = price.replace('$','')
        return(x)
    else:
        return price

joined_data['Price'] = joined_data['Price'].apply(remove).astype(float)
```

```python
#Function to convert Size column to integer, in kilobytes

def convert_bytes(size):
    x = 0
    if 'M' in size:
        x = size[:-1]
        x = float(x) * 1000
        return(x)
    elif 'k' in size:
        x = size[:-1]
        return float(x)
    else:
        return float(x)

joined_data["Size"] = joined_data["Size"].apply(convert_bytes) 
joined_data
```

```python
#Function to remove + from installs
def remove_sign(data):
    x = 0
    x = data.replace(',','')
    x = x[:-1]
    return x

joined_data['Installs'] = joined_data['Installs'].map(remove_sign)
joined_data['Installs'] = joined_data['Installs'].astype(int)
```

```python
#Function to change Type column
def change_type(type):
    if type == "Free":
        return 1
    else:
        return 0

joined_data['Type'] = joined_data['Type'].apply(change_type)
joined_data
```

```python
# Apps Category getting the highest number of Sentiments
joined_data.groupby('Category').size().nlargest(5).reset_index(name='frequency')
```

```python
# Using pivot table on Joined Table to find out the Positive,Negative and Neutral Sentiments Counts for each Category
joined_data.pivot_table(columns='Sentiment',index='Category',values='Rating',aggfunc='count')
```

Executed the Correlation Matrix to find the corrleation between different independent variables in the joined dataset.

```python
f, ax = plt.subplots(1, 1, figsize=(10, 8))
correlation_succ = joined_data.corr()
sns.heatmap(correlation_succ, cmap="coolwarm_r", annot=True, ax=ax)

ax.set_ylim(len(joined_data.columns)-2, 0)
plt.show()
```

```python
joined_data.corr()
```

Using One-Hot Encoding to transform the categorical variables. In our joined data we have 3 categorical variables which are Category,Genres and Sentiment.

```python
## selecting only the categorical variables
cat_joined_data= joined_data[['Category','Genres','Sentiment']]
```

```python
from sklearn.preprocessing import OneHotEncoder
#Create an Encoder
encoder=OneHotEncoder(sparse=False)
#Fit the encoder to the columns we want to transform
encoder.fit(cat_joined_data)
```

```python
#Transform these columns
cat_cols_1hot=encoder.transform(cat_joined_data)
cat_cols_1hot
```

```python
#Convert the results back to dataframe
cat_cols_1hot_df=pd.DataFrame(cat_cols_1hot,columns=encoder.get_feature_names())
cat_cols_1hot_df
```

```python
# Append them back to the original
joined_data=pd.concat([joined_data.drop(['Category','Genres','Sentiment'],axis=1),cat_cols_1hot_df],axis=1)
```

After our joined dataset is merged together with Categorical Varibles, we are going to split the data into train and test Dataset.
For Predictor Variables, we are going to drop 'Rating' since it is our target variable and also we are going to drop 'App' since the categorical variable has lot of distinct values.

```python
X=joined_data.drop(['Rating','App'], axis=1)
```

```python
X.info()
```

```python
X.columns
```

We include the target variable which is 'Rating'.

```python
y = joined_data['Rating']
```

The dataset is splitted in the ratio of 70:30 with respect to train and test dataset respectively.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
print (X_train.shape, y_train.shape)
print (X_test.shape, y_test.shape)
```

So since the variables in our model has been deduced we are going to perform regression models.

```python
from sklearn.linear_model import Lasso,LinearRegression,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.metrics import confusion_matrix,accuracy_score
```

Linear Regression: We are going to create our model based on the train dataset and then test the accuracy based on test dataset.

```python
# Building a Linear Regression
Lin_R=LinearRegression()
lin_model = Lin_R.fit(X_train,y_train)
y_predict=lin_model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict})
df.head(15)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_predict))  
print('Mean Squared Error:', mean_squared_error(y_test, y_predict))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_predict)))
```

```python
Lin_R.score(X_test, y_test)
```

So we observed that the R-square value is 0.267.

Next model which we are going to build is the Lasso Regression using the test and train dataset mechanism.

```python
# Building a Lasso Regression
Lasso_R=Lasso()
lasso_model = Lasso_R.fit(X_train,y_train)
y_predict_lasso=lasso_model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict_lasso})
df.head(15)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_predict_lasso))  
print('Mean Squared Error:', mean_squared_error(y_test, y_predict_lasso))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_predict_lasso)))
```

```python
Lasso_R.score(X_test, y_test)
```

So we observed that the R-square value using Lasso Regression was very less and hence we would try using Ridge regression.

Next model which we are going to build is the Ridge regression using the test and train dataset mechanism.

```python
# Building a Ridge Regression
Ridge_R=Ridge()
Ridge_R.fit(X_train,y_train)
y_predict_Ridge=Ridge_R.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict_Ridge})
df.head(15)
```

```python
Ridge_R.score(X_test, y_test)
```

So we observed that the R-square value using Lasso Regression closer to the Linear Regression.
As the target variable is a continuous variable we would not be able to use the Logistic Regression model.
Next, we would test our dataset using non-Linear models.
First model under non-Linear models is the Decision Tree Regressor.

```python
# Decision Tree
Dt = DecisionTreeRegressor()
dt_model = Dt.fit(X_train,y_train)
y_predict_dt=dt_model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict_dt})
df.head(15)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_predict_dt))  
print('Mean Squared Error:', mean_squared_error(y_test, y_predict_dt))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_predict_dt)))
```

```python
Dt.score(X_test, y_test)
```

Thus we observed that the R-square is very high which gives us the indication that our model is a non-linear model.
We would try using Random Forest to further illustrate this point.

```python
# Random Forest
Rf=RandomForestRegressor()
Rf_model = Rf.fit(X_train,y_train)
y_predict_rf=Rf_model.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_predict_rf})
df.head(15)
```

```python
print('Mean Absolute Error:', mean_absolute_error(y_test, y_predict_rf))  
print('Mean Squared Error:', mean_squared_error(y_test, y_predict_rf))  
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(y_test, y_predict_rf)))
```

```python
Rf.score(X_test, y_test)
```

The R-square value of the Random Forest Regression also proves the fact that our model is a non-linear one.

We are going to further test our model using Neural Network and we are going to use Keras.

```python
# Keras 
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
```

```python
# define the keras model
model = Sequential()

#add model layers
model.add(Dense(10, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dense(10, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
from keras.callbacks import EarlyStopping
#set early stopping monitor so the model stops training when it won't improve anymore
early_stopping_monitor = EarlyStopping(patience=3)
#train model
model.fit(X_train, y_train, validation_split=0.2, epochs=30, callbacks=[early_stopping_monitor])
```

```python
#example on how to use our newly trained model on how to make predictions on unseen data (we will pretend our new data is saved in a dataframe called 'test_X').
test_y_predictions = model.predict(X_test)
test_y_predictions
```

```python
model.evaluate(X_test,y_test)
```
